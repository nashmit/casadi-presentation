## Automatic differentiation (AD)

Consider an ode:

\begin{equation}
\dot{p} = (1 - q^2)p-q+u
\end{equation}
\begin{equation}
\dot{q} = p
\end{equation}
\begin{equation}
\dot{c} = p^2+q^2+u^2
\end{equation}

```python
t = SX.sym("t")    # time
u = SX.sym("u")    # control
p = SX.sym("p")
q = SX.sym("q")
c = SX.sym("c")
x = vertcat(p,q,c) # state
```

```python
x
```

```python
ode = vertcat((1 - q**2)*p - q + u, p, p**2+q**2+u**2)
print(ode)
```

```python
print(ode.shape)
```

```python
J = jacobian(ode,x)
```

```python
print(J)
```

```python
sp=J.sparsity()
sp
```

```python
sp.is_symmetric()
```

```python
J.sparsity().spy()
```

```python
f = Function("f",[t,u,x],[ode])

ffwd = f.forward(1)

fadj = f.reverse(1)

# side-by-side printing
print('{:*^24} || {:*^28} || {:*^28}'.format("f","ffwd","fadj"))
def short(f):
    import re
    return re.sub(r", a\.k\.a\. \"(\w+)\"",r". \1",str(f).replace(", No description available","").replace("Input ","").replace("Output ",""))
for l in zip(short(f).split("\n"),short(ffwd).split("\n"),short(fadj).split("\n")):
  print('{:<24} || {:<28} || {:<28}'.format(*l))
```

Performing **forward sweeps** gives the **columns** of J

```python
print(I)
```

```python
for i in range(3):
  print(ffwd(t,u,x, ode, 0,0,I[:,i]))
```

```python
print(J)
```

Performing **adjoint sweeps** gives the **rows** of J

```python
for i in range(3):
  print(fadj(t,u,x, ode, I[:,i])[2])
```

Often, you can do better than slicing with unit vectors

## Integrators

$\dot{x}=f(x,u,t)$ with $x = [p,q,c]^T$

```python
f = {'x':x,'t':t,'p':u,'ode':ode}
```

Construct an integrating block $x_{k+1} = \Phi(f;\Delta t;x_k,u_k)$

```python
tf = 10.0
N = 20
dt = tf/N
```

```python
Phi = integrator("Phi","cvodes",f,{"tf":dt})
print(Phi)
```

```python
x0 = DM([0,1,0])

print(Phi(x0=x0))
```

```python
x = x0
xs = [x]

for i in range(N):
  x = Phi(x0=x)["xf"]
  
  xs.append(x)
```

```python
xs
```

```python
horzcat(*xs).shape
```

```python
horzcat(*xs)
```

```python
horzcat(*xs).T
```

```python
plot(horzcat(*xs).T)
legend(["p","q","c"])
show()
```

## Matrix expression (MX) graphs

> Note 4: this is what makes CasADi stand out among AD tools

Recall

```python
n = 3

A = SX.sym("A",n,n)
B = SX.sym("B",n,n)
C = mtimes(A,B) # A@B
print(C.shape)
print(symvar(C))
print(C)
dotdraw(C,direction='BT')
```

What if you **don't want** to expand into scalar operations?  ( avoid $O(n^3)$ storage)

```python
A = MX.sym("A",n,n)
B = MX.sym("B",n,n)
C = A@B
print(symvar(C))
print(C.shape)
print(C)
dotdraw(C,direction='BT')
```

What if you **cannot** expand into matrix operations?  ( numerical algorithm )

```python
C = solve(A,B)
print(symvar(C))
print(C.shape)
print(C)
dotdraw(C,direction='BT')
```

### If else statement inside a CasADi expression

```python
#maximum or if a>b return a
#(a+b + abs(a-b))/2

a = SX.sym('a')
b = SX.sym('b')

c = (a + b + fabs(a-b))/2
print(c)
print(jacobian(c,vertcat(a,b)))
print(gradient(c,vertcat(a,b)))
print('--')
print(jacobian(if_else(a>b,a,b),vertcat(a,b)) )
print(gradient(if_else(a>b,a,b),vertcat(a,b)) )
print('--')
print(gradient(fabs(a-b),vertcat(a,b)))
print('--')
print(jacobian(fabs(a),vertcat(a,b)))
print('--')
print(gradient(sign(a-b),vertcat(a,b)))
```

```python
import numpy as np


x = SX.sym('x',1)
t = SX.sym('t',1)

xdot1 = 0
xdot2 = 1
xdot3 = x

t0 = 0
tf = 3

def func():
    #print(xdot2)
    return xdot2

#xdot = if_else( t<1, xdot1, if_else( t<2, xdot2, xdot3 ) ) # change dynamics
xdot = if_else(t<1, xdot1, func())

# don't know how to change initial condition when an event is happening ( alter current state 'x' )


ode = {'x':x, 't':t, 'ode':xdot }
time = np.insert(np.linspace(t0, tf, 20), t0, 0)

I = integrator('I','cvodes', ode, {'grid': time})

out = I(x0=1)['xf']
print(out)

plot(out.T)
show()
```

### Vectorizaiton: "map" integrator example

```python
#Lotka–Volterra
def DefineOneShootingNodForDAE_LV(t0, tf, NrEvaluationPoints = 40):
    x0 = SX.sym('x0');
    x1 = SX.sym('x1');
    x = vertcat(x0, x1);
    p00 = SX.sym('p00');
    p01 = SX.sym('p01');
    p10 = SX.sym('p10');
    p11 = SX.sym('p11');
    p = vertcat(horzcat(p00, p01), horzcat(p10, p11));
    xdot = vertcat(p00 * x0 - p01 * x0 * x1, p10 * x0 * x1 - p11 * x1);
    ode = {'x': x, 'p': reshape(p, 4, 1), 'ode': xdot};

    time = np.insert(np.linspace(t0, tf, NrEvaluationPoints), t0, 0)
    I = integrator('I', 'cvodes', ode, {'t0': t0, 'tf': tf});
    I_plot = integrator('I_plot', 'cvodes', ode, {'grid': time});

    # returns [integrator][Nr equations][Nr parameters]
    return I, 2, 4, I_plot;
```

```python
def BuildCasADiExpFromIntegrator(NumberShootingNodes, intg, SizeOfx0, SizeOfp):
    p = MX.sym('p', SizeOfp);

    #map(N,"thread",2)
    F = intg.map(NumberShootingNodes)
    print(F)
    print("-----------")
    Sn = MX.sym('Sn', SizeOfx0, NumberShootingNodes)
    intg_exp = F(x0=Sn, p=repmat(p, 1, NumberShootingNodes))

    # return [symbolic input variables] and [symbolic integrator as expression]
    return Sn, p, intg_exp
```

```python
t0=0
tf=10
NumberShootingNodes=8

intg, SizeOfx0, SizeOfp, intg_plot = DefineOneShootingNodForDAE_LV(t0=0, tf=tf / NumberShootingNodes)
print(intg)
print("------")

Sn, p, intg_exp = BuildCasADiExpFromIntegrator(NumberShootingNodes=NumberShootingNodes,intg=intg,SizeOfx0=SizeOfx0, SizeOfp=SizeOfp)

print(Sn)
print(Sn.shape)
print(p)
print(p.shape)
print(intg_exp['xf'])
print(symvar(intg_exp['xf']))
```
